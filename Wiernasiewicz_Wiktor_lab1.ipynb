# Zadanie 3

W trzecim zadaniu zdefiniowałem funkcję f(x)=x^2+5, a następnie sporządziłem wykres dla argumentów z danego przedziału.

```python
#zaimportowanie niezbędnych bibliotek
import numpy as np
from matplotlib import pyplot as plt
```

```python
def my_function(x):
    return x**2+5

#zadane przedziały funkcji
x1 = np.linspace(-1,1)
x2 = np.linspace(-6,6)
x3 = np.linspace(0,5)
```

```python
#przedstawienie wykresów funkcji dla danych argumentów
plt.plot(x1, my_function(x1),label='<-1,1>')
plt.title("f(x)=x^2 + 5 dla przedziału <-1,1>")
plt.xlabel("Argumenty")
plt.ylabel("Wartości funkcji")
plt.legend("f(x)=x^2 + 5")
plt.grid()
plt.show()
```

```python
plt.plot(x2, my_function(x2),label='<-6,6>')
plt.title("f(x)=x^2 + 5 dla przedziału <-6,6>")
plt.xlabel("Argumenty")
plt.ylabel("Wartości funkcji")
plt.legend("f(x)=x^2 + 5")
plt.grid()
plt.show()
```

```python
plt.plot(x3, my_function(x3),label='<0,5>')
plt.title("f(x)=x^2 + 5 dla przedziału <0,5>")
plt.xlabel("Argumenty")
plt.ylabel("Wartości funkcji")
plt.legend("f(x)=x^2 + 5")
plt.grid()
plt.show()
```
